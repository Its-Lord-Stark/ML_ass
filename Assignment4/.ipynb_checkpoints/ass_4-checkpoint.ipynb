```python
####Gradient Descent for local minima for function (x+3)^2

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Initialize all parameters

cur_x = 2
max_iter = 10000
iters=0
prev_step_size=1
learning_rate=0.01
precesion = 0.000001
points = [cur_x]
df = lambda x: 2*(x+3)
```

```python
while(prev_step_size>precesion and iters<max_iter):
    prev_x = cur_x
    cur_x = cur_x-learning_rate*df(prev_x)
    prev_step_size = abs(cur_x-prev_x)
    iters = iters+1
    points.append(cur_x)
    print("Iterations", iters, "\nx value is", cur_x)
    
print("Local minima is", cur_x)
    
```

```python
plt.plot(points)
```
