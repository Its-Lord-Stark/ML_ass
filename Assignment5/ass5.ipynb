```python
def rem_outlier(data):
    Q1 = np.percentile(data,25)
    Q2 = np.percentile(data,50)
    Q3 = np.percentile(data,75)
    IQR = Q3-Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    return(lower_bound,upper_bound)
```

```python
lower_bound,upper_bound = rem_outlier(df.Pregnancies)
lower_bound,upper_bound
```

```python
df = df[(df.Pregnancies>=lower_bound) & (df.Pregnancies<=upper_bound)]
```

```python
y = df.Outcome
y
```

```python
x = df.drop(["Outcome", "SkinThickness"], axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,classification_report,recall_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
cnf = confusion_matrix(y_pred,y_test)
cnf
```

```python
prec = precision_score(y_pred,y_test)
prec
```

```python
rec = recall_score(y_pred,y_test)
rec
```

```python
print(classification_report(y_test,y_pred))
```
