```python
df1 = df[["PRICEEACH","MSRP"]]

df1.shape
```

```python
df1.isna().sum()
```

```python
from sklearn.cluster import KMeans

inertia = []

for i in range(1,11):
    kmean = KMeans(n_clusters = i)
    kmean.fit(df1)
    inertia.append(kmean.inertia_)
```

```python
kmean = KMeans(n_clusters=2)
pred = kmean.fit_predict(df1)
```

```python
sns.scatterplot(x=df1["PRICEEACH"],y=df1["MSRP"],hue=pred)
```
