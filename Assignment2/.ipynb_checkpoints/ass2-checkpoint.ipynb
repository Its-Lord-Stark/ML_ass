```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("C:/Users/BFLCOMP01/Desktop/ML/Assignment2/emails.csv")
```

```python
df.drop(columns='Email No.', inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df.Prediction.unique()
```

```python
df.Prediction = df.Prediction.replace({0:'Not spam', 1:'Spam'})
```

```python
df
```

```python
X = df.drop(columns='Prediction')
Y = df.Prediction
```

```python
X.columns
```

```python
Y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
KN = KNeighborsClassifier
knn = KN(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python
M = metrics.accuracy_score(y_test,y_pred)
print(M)
```

```python
C = metrics.confusion_matrix(y_test,y_pred)
print(C)
```

```python
model = SVC(C=1)
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
M1 = metrics.accuracy_score(y_test,y_predict)
print(M1)
```

```python
C1 = metrics.confusion_matrix(y_test,y_predict)
print(C1)
```
